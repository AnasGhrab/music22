```python
%matplotlib inline
import timeside
from timeside.core import get_processor
```

```python
audio_source = '/Users/anas/AUDIO/Barraq/P1.wav'
FileDecoder = get_processor('file_decoder')
decoder = FileDecoder(uri=audio_source)
```

```python
level = get_processor('level')()
```

```python
(decoder | level).run()
```

```python
level.results.keys()
```

```python
max = level.results['level.max']
print max.data
```

```python
rms = level.results['level.rms']
print rms.data
```

```python
%pylab inline

from __future__ import division
import matplotlib.pylab as pylab
pylab.rcParams['figure.figsize'] = 16, 12  # that's default image size for this interactive session
import timeside
from timeside.core import get_processor

import matplotlib.pyplot as plt
import numpy as np
import sys

audiofile = '/Users/anas/AUDIO/Barraq/P1.wav'

# normal
file_decoder = get_processor('file_decoder')(audiofile, start=0)
aubio_pitch = get_processor('aubio_pitch')()
aubio_temporal = get_processor('aubio_temporal')()
specgram_ = get_processor('spectrogram_analyzer')()
waveform = get_processor('waveform_analyzer')()
#g  =  timeside.grapher.Spectrogram()

pipe = (file_decoder | aubio_pitch | aubio_temporal | specgram_ | waveform)
print pipe
```

```python
pipe.run()

print pipe.results.keys()
```

```python
res_pitch = aubio_pitch.results['aubio_pitch.pitch']
plt.plot(res_pitch.time, res_pitch.data)

plt.title('Spectrogram + Aubio pitch + Aubio beat')
plt.grid()
```
